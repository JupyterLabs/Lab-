# Image Import and Rotates

PIL:python imaging library

Import Image: First of all upload image in Jupyter notebook,and write code below in cell (in code type path of image which you want to import) and Run


```python
from PIL import Image as PIL_Image

orig_img_path = '/home/xilinx/jupyter_notebooks/Monal.jpg'
img = PIL_Image.open(orig_img_path)
img
```

### RGB To Grayscale conversion

Image convert for RGB to gray scale use below code,
"L" use for grayscale ,bw : binary image

```python
bw_img = img.convert("L")
bw_img
```

### Image roatation 45 angle

use "img.rotate(angle)" for rotate image to specific angle.

```python
rot_img = img.rotate(45)
rot_img
```

### Image roatation 180 angle

```python
rot_img = img.rotate(180)
rot_img
```
